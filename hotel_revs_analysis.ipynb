```python
# running analysis on a select set of columns of importance to the analysis of the dataset.
selected_cols = [
    'Hotel_Name',
    'Reviewer_Score',
    'Total_Number_of_Reviews', 
    'lat', 
    'lng'
]

hotel_freqs_df = df[selected_cols].copy()
```

```python
hotel_freqs_df.head(5)
```

```python
#Creating a unique identifier for each hotel based on name and location (latitude and longitude)
# to avoid confusion between hotels with similar names in different locations. 
# This will help in accurately analyzing the data related to each specific hotel.
hotel_freqs_df['Hotel_ID'] = (
    hotel_freqs_df['Hotel_Name'].astype(str)
    + "_"
    + hotel_freqs_df['lat'].round(3).astype(str)
    + "_"
    + hotel_freqs_df['lng'].round(3).astype(str)
)
```

```python
#Calculating the total number of reviews found for each hotel by grouping the data by 'Hotel_ID' and
# counting the occurrences of each unique hotel identifier. 
# This will provide insight into how many reviews are associated with each specific hotel in the dataset.
# No guarantee to validate 'Total_Nuumber_of_Reviews' column in original dataset,
# so we will calculate our own 'Total_Reviews_Found' column 
# based on the actual number of reviews found in the dataset for each hotel.
hotel_freqs_df['Total_Reviews_Found'] = (
    hotel_freqs_df.groupby('Hotel_ID')
    .Hotel_ID
    .transform('size')
)
```

```python
#dropping duplicates based on 'Hotel_ID'.
hotel_summary = hotel_freqs_df.drop_duplicates(
    subset=['Hotel_ID']
).copy()
```

```python
# define a function that takes a row and performs some calculation with it
def get_difference_review_avg(row):
  return row["Average_Score"] - row["Calc_Average_Score"]

df['Calc_Average_Score'] = round(df.groupby('Hotel_Name').Reviewer_Score.transform('mean'), 1)

# Add a new column with the difference between the two average scores
df["Average_Score_Difference"] = df.apply(get_difference_review_avg, axis = 1)

# Create a df without all the duplicates of Hotel_Name (so only 1 row per hotel)
review_scores_df = df.drop_duplicates(subset = ["Hotel_Name"])

# Sort the dataframe to find the lowest and highest average score difference
review_scores_df = review_scores_df.sort_values(by=["Average_Score_Difference"])

display(review_scores_df[["Average_Score_Difference", "Average_Score", "Calc_Average_Score", "Hotel_Name"]])
```

## After exploration it was found only the following cities and countries were in the data set -
Amsterdam, Netherlands

Barcelona, Spain

London, United Kingdom

Milan, Italy

Paris, France

Vienna, Austria

```python
def replace_address(row):
    if "Netherlands" in row["Hotel_Address"]:
        return "Amsterdam, Netherlands"
    elif "Barcelona" in row["Hotel_Address"]:
        return "Barcelona, Spain"
    elif "United Kingdom" in row["Hotel_Address"]:
        return "London, United Kingdom"
    elif "Milan" in row["Hotel_Address"]:        
        return "Milan, Italy"
    elif "France" in row["Hotel_Address"]:
        return "Paris, France"
    elif "Vienna" in row["Hotel_Address"]:
        return "Vienna, Austria" 

# Replace all the addresses with a shortened, more useful form
df["Hotel_Address"] = df.apply(replace_address, axis = 1)
# The sum of the value_counts() should add up to the total number of reviews
print(df["Hotel_Address"].value_counts())
```
